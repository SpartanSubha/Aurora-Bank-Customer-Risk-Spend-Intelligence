```python
# Remove $ sign before the amount in users_data

for col in ['per_capita_income',	'yearly_income',	'total_debt']:
  udf[col] = udf[col].str.replace('[\$,]','',regex='True').astype(int)

udf.head()
```

```python
cdf['credit_limit'] = cdf['credit_limit'].str.replace('[\$,]', '', regex='True').astype(int)
cdf.head()
```

```python
# Dropping Latitude columns having absurd values and replace with Null

udf.loc[pd.to_numeric(udf['latitude'], errors='coerce').isna(), 'latitude'] = None
```

```python
udf['latitude'] = udf['latitude'].astype(float)
```
